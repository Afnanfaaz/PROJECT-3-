# Dataset Structure Description

The dataset for this analysis is an air traffic passenger statistics dataset. Below is a description of each column in the dataset:

1. **Activity Period**: Numeric representation of the year and month when the activity was recorded (format: YYYYMM).
2. **Activity Period Start Date**: The starting date of the activity period (format: YYYY/MM/DD).
3. **Operating Airline**: Name of the airline operating the flight.
4. **Operating Airline IATA Code**: International Air Transport Association (IATA) code for the operating airline.
5. **Published Airline**: Name of the airline under which the flight is marketed.
6. **Published Airline IATA Code**: IATA code for the published airline.
7. **GEO Summary**: Geographical categorization of the flight (e.g., Domestic, International).
8. **GEO Region**: Specific geographical region of the airline operation (e.g., Europe, Asia).
9. **Activity Type Code**: Type of activity recorded (e.g., Deplaned, Enplaned, Thru / Transit).
10. **Price Category Code**: Category of the flight pricing (e.g., Low Fare, Other).
11. **Terminal**: Terminal from which the flight operated.
12. **Boarding Area**: Specific boarding area within the terminal.
13. **Passenger Count**: Count of passengers associated with the recorded activity.
14. **data_as_of**: The timestamp indicating when the data was last updated.
15. **data_loaded_at**: The timestamp indicating when the data was loaded into the dataset.

This dataset provides comprehensive information on air traffic movements, including passenger counts, airline operations, and geographical distribution. It serves as a rich source for analyzing trends and patterns in air travel.

# **Airline Comparison: Passenger Counts for Domestic and International Flights**

For this analysis, I'll compare airlines based on their passenger counts, distinguishing between domestic and international flights. I'll create separate plots for domestic and international airlines to visualize which airlines have the highest passenger traffic in each category. ​​



```python
# Filtering data for Domestic and International flights
domestic_data = air_traffic_data[air_traffic_data['GEO Summary'] == 'Domestic']
international_data = air_traffic_data[air_traffic_data['GEO Summary'] == 'International']

# Aggregating passenger count by airline for Domestic and International flights
domestic_airline_passengers = domestic_data.groupby('Operating Airline')['Passenger Count'].sum().sort_values(ascending=False)
international_airline_passengers = international_data.groupby('Operating Airline')['Passenger Count'].sum().sort_values(ascending=False)

# Plotting for Domestic flights
plt.figure(figsize=(15, 7))
sns.barplot(x=domestic_airline_passengers.head(10).values, y=domestic_airline_passengers.head(10).index)
plt.title('Top 10 Airlines by Passenger Count - Domestic Flights')
plt.xlabel('Total Passenger Count')
plt.ylabel('Airlines')
plt.show()

# Plotting for International flights
plt.figure(figsize=(15, 7))
sns.barplot(x=international_airline_passengers.head(10).values, y=international_airline_passengers.head(10).index)
plt.title('Top 10 Airlines by Passenger Count - International Flights')
plt.xlabel('Total Passenger Count')
plt.ylabel('Airlines')
plt.show()
```

The bar plots above display the top 10 airlines based on the total passenger count, separately for domestic and international flights. These visualizations help in identifying which airlines dominate in terms of passenger traffic in each category.

The bar charts provide a clear comparison of the passenger count handled by different terminals and across various geographical regions.This bar chart ranks the top 10 airlines based on passenger counts for domestic flights. United Airlines holds the top position, followed closely by American Airlines and SkyWest Airlines. It's notable that United Airlines also has a separate entry for data pre-July 2013, indicating a structural or reporting change at that time.

# **Monthly and Yearly Trends Analysis**

```python
# Extract month and year from 'Activity Period Start Date'
air_traffic_data['Month'] = pd.to_datetime(air_traffic_data['Activity Period Start Date']).dt.month

# Grouping data by Year and Month
monthly_data = air_traffic_data.groupby(['Year', 'Month'])['Passenger Count'].sum().reset_index()

# Creating a pivot table for year-over-year comparison
pivot_table = monthly_data.pivot("Month", "Year", "Passenger Count")

# Plotting the pivot table as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="coolwarm")
plt.title('Monthly Passenger Count Trends Year-over-Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

The heatmap above illustrates the monthly passenger count trends year-over-year. Each cell represents the total number of passengers for a given month and year, allowing us to observe both seasonal and annual patterns. The heatmap showcases monthly passenger count trends on a year-over-year basis. The intensity of the colors represents the volume of passengers, with warmer colors indicating higher numbers. This visualization helps identify seasonal patterns or particular months with significant deviations from the norm.

# **Detailed Airline and Region Analysis**

Analyze which airlines have the highest growth in passenger counts year-over-year, segmented by geographic region.

```python
# Grouping data by airline, region, and year
airline_region_yearly = air_traffic_data.groupby(['Operating Airline', 'GEO Region', 'Year'])['Passenger Count'].sum().reset_index()

# Calculating Year-over-Year Growth for each airline in each region
airline_region_yearly['YoY Growth'] = airline_region_yearly.groupby(['Operating Airline', 'GEO Region'])['Passenger Count'].pct_change() * 100

# Filtering out the first year for each group as YoY growth can't be calculated for the first year
airline_region_yearly_filtered = airline_region_yearly.dropna(subset=['YoY Growth'])

# To display: Selecting a few airlines and regions for demonstration
selected_airlines = airline_region_yearly_filtered['Operating Airline'].unique()[:5]
selected_data = airline_region_yearly_filtered[airline_region_yearly_filtered['Operating Airline'].isin(selected_airlines)]

# Plotting
plt.figure(figsize=(15, 10))
sns.lineplot(x='Year', y='YoY Growth', hue='Operating Airline', style='GEO Region', data=selected_data)
plt.title('Year-over-Year Growth in Passenger Count by Airline and Region')
plt.xlabel('Year')
plt.ylabel('YoY Growth (%)')
plt.legend(title='Airline and Region')
plt.show()
```

Year-over-Year Growth in Passenger Count by Airline and Region The line chart displays year-over-year growth in passenger counts, segmented by airlines and regions. Notable spikes and variances in growth percentages can be observed, which could be attributed to various factors such as seasonal demand, airline promotions, or external events affecting travel.
